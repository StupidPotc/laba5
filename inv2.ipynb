<h3>На диаграмме представлены планеты Солнечной системы, укажите какая из планет расположена Ближе от Сатурна к Солнцу примерно в 10 раз?  <h3>

```python
import matplotlib.pyplot as plt
%matplotlib inline
fruits = ["Нептун", "Уран", "Сатурн", "Юпитер", "Марс", "Земля", " Венера", " Меркурий"]
counts = [4500, 2900, 1400, 780, 230, 150, 108, 58]
plt.bar(fruits, counts)
plt.title("Диаграмма планет по расстонию до Солнца")
plt.xlabel("Название планеты")
plt.ylabel("расстояние до Солнца, млн км.")
```

<h4>Ответ: Земля<h4>
