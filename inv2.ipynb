<h3>На диагрмме представлены крупнейшие страны мира по численности населения (млн.чел).Численность населения какого государства примерно в 6 раз больше численности населения Бразилии?<h3>

```python
import matplotlib.pyplot as plt
%matplotlib inline
fruits = ["Китай", "Индия", "США", "Пакистан", "    Бразилия", "   Россия", " Япония"]
counts = [1300, 1000, 300, 200, 190, 150, 120]
plt.bar(fruits, counts)
plt.title("Диаграмма стран мира по численности населения")
plt.xlabel("Название стран")
plt.ylabel("численности населения, млн чел.")
```

<h4>Ответ: Индия<h4>
