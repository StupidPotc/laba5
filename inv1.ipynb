<h2> Задача по физике <h2>

<h4>Расстояние (S) между городами М и К = 250 км. Одновременно из обоих городов навстречу друг другу выезжают автомашины. Машина из города М движется со скоростью = 60 км/ч, из города К — со скоростью ν2 = 40 км/ч. Построить график зависимости пути от времени для каждой из машин и по ним определить место встречи и время их движения до встречи.<h4>

<h5> Решение <h5>
    Движение происходит вдоль оси Х между точкой 0 и точкой 300 , согласно формуле
    

x1(t) = v1*t


x2(t) = s - v2*t

```python
import matplotlib.pyplot as plt
plt.plot([0, 50,100, 150, 200, 250, 300])
plt.plot([250, 200, 150, 100, 50,0])
plt.ylabel('x, км', fontsize=12, color='blue')
plt.xlabel('t, ч', fontsize=12, color='blue')
plt.text(4.8, 270, 'X1(t)')
plt.text(4.8, 35, 'X2(t)')
plt.grid()
plt.show()
```

<h5>Графики этих функицй изображены на рисунке. В момент втречи координаты машин совпадают<h5>

<h6>x1(t)=x2(t0)=x0<h6>

<h5>Проекция точки пересечения на ось T показывает время t0, когда это произошло.<h5>
